## Data

```python
x = np.random.normal(0, 1, size=100)
y = np.random.normal(0, 1, size=100)
z = np.random.normal(0, 1, size=100)
c = np.random.normal(0, 1, size=100)
```

## Plots Ex.1.

### v.1.

```python
start = time.time()
# ...
plt.figure(figsize=(6,6), dpi=100)

# ...
plt.scatter(x, y, s=z*500, c=c, alpha=0.5)

# ...
plt.colorbar()
plt.grid()

plt.show();
stop = time.time()

print('===> Time : {} sec. <==='.format(stop - start))
```

### v.2.

```python
start = time.time()
# ...
plt.figure(figsize=(6,6), dpi=100)

# ...
plt.plot(x, y, "o", markersize=10, c='r', alpha=0.5 ) ## Not Dynamic <<========

# ...
# plt.colorbar()
plt.grid()

plt.show();
stop = time.time()

print('===> Time : {} sec. <==='.format(stop - start))
```
