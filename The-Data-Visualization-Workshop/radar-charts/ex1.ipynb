## Data

```python
# Sample data
# Attributes: Efficiency, Quality, Commitment, Responsible Conduct, Cooperation
data = pd.DataFrame({
    'Employee': ['Alex', 'Alice', 'Chris', 'Jennifer'],
    'Efficiency': [5, 4, 4, 3,],
    'Quality': [5, 5, 3, 3],
    'Commitment': [5, 4, 4, 4],
    'Responsible Conduct': [4, 4, 4, 3],
    'Cooperation': [4, 3, 4, 5]
})
```

```python
data
```

## Plot

```python
attributes = list(data.columns[1:])
attributes
```

```python
values = data[['Efficiency','Quality', 'Commitment', 'Responsible Conduct', 'Cooperation']].values
values
```

```python
employees = data.Employee.values
employees
```

```python
angles = [n / float(len(attributes)) * 2 * np.pi for n in range(len(attributes))]
# Close the plot
angles += angles[:1]
angles
```

```python
values = np.asarray(values)
values = np.concatenate([values, values[:, 0:1]], axis=1)
values
```

```python
# ...
plt.figure(figsize=(8, 8), dpi=150)

# ...
for i in range(values.shape[0]):
    # ..
    ax = plt.subplot(2, 2, i + 1, polar=True)
    ax.plot(angles, values[i])
    ax.fill(angles, values[i], 'b', alpha=0.1)

    # ..
    ax.set_yticks([1, 2, 3, 4, 5]) # yticks ...
    ax.set_xticks(angles) # xticks ...
    ax.set_xticklabels(attributes) # xticklabels ...

    # ..
    ax.set_title(employees[i], fontsize=14, color='r')


# ...
plt.tight_layout()
plt.show();
```
