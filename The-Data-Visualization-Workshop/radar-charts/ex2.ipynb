```python
import numpy as np
import pandas as pd
import os
import time

import matplotlib.pyplot as plt
%matplotlib inline
```

## Data

```python
# Sample data
# Attributes: Efficiency, Quality, Commitment, Responsible Conduct, Cooperation
df = pd.DataFrame({
    'Employee': ['Alex', 'Alice', 'Chris', 'Jennifer'],
    'Efficiency': [5, 4, 4, 3,],
    'Quality': [5, 5, 3, 3],
    'Commitment': [5, 4, 4, 4],
    'Responsible Conduct': [4, 4, 4, 3],
    'Cooperation': [4, 3, 4, 5]
})
```

```python
df
```

## Plot

```python
attr = ['Efficiency','Quality', 'Commitment', 'Responsible Conduct', 'Cooperation']
attr
```

```python
# angels ...
N = len(attr)

angles = [a / float(N) * 2 * np.pi for a in range(N)]
angles += angles[:1]
angles
```

```python
# values ...
values = df[attr].values
values = np.concatenate([values, values[:, 0:1]], axis=1)

values
```

```python
val_1 = values[0]
val_1
```

```python
val_3 = values[3]
val_3
```

```python
# ...
plt.figure(figsize=(4, 4), dpi=100)

# ...
ax = plt.subplot(111, polar=True)
ax.plot(angles, val_1, linewidth=1, linestyle='solid', label="Alex")
ax.fill(angles, val_1, 'b', alpha=0.1)

ax.plot(angles, val_3, linewidth=1, linestyle='solid', label="Chris")
ax.fill(angles, val_3, 'r', alpha=0.1)

# ...
ax.set_yticks([1, 2, 3, 4, 5]) # yticks ...
ax.set_xticks(angles) # xticks ...
ax.set_xticklabels(attr) # xticklabels ...

plt.title('Rader Chart')
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show();
```
