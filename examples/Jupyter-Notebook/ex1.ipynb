```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,4*np.pi,0.1) 
y = np.cos(x)
```

```python
plt.figure(figsize=(12,6))
plt.plot(
    x,
    y,
    label='coswave',
    color='green',
    marker='^',
    linestyle='-',
    linewidth = 2,
    markerfacecolor='red',
    markersize=7
)

plt.show();
```

```python
plt.figure(figsize=(12,6))
plt.plot(
    x,
    y,
    label='coswave',
    color='green',
    marker='^',
    linestyle='-.',
    linewidth = 2,
    markerfacecolor='red',
    markersize=7
)

plt.show();
```

```python
plt.figure(figsize=(12,6))
plt.plot(
    x,
    y,
    label='coswave',
    color='green',
    marker='o',
    linestyle='-.',
    linewidth = 2,
    markerfacecolor='green',
    markersize=7
)

plt.show();
```
