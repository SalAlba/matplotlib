```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
a = np.arange(1., 5., 0.2)
```

### Basic Quickstart Guide

```python
fig1, f1_axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True);
```

```python
# Indexes start from 0

fig2 = plt.figure(constrained_layout=True)

spec2 = gridspec.GridSpec(nrows=4, ncols=3, figure=fig2)

f2_ax1 = fig2.add_subplot(spec2[0, 0])
f2_ax1.set_title('[0, 0]') # set_title()
plt.plot(a, a)
plt.plot(a, a**2)

f2_ax2 = fig2.add_subplot(spec2[0, 2])
f2_ax2.set_title('[0, 2]') # set_title()
plt.plot(a, a)
plt.plot(a, a**2)


f2_ax3 = fig2.add_subplot(spec2[1:3, 0])
plt.title('[1:3, 0]') # plt.title()

f2_ax4 = fig2.add_subplot(spec2[1, 1:])
plt.title('[1, 1:]') # plt.title()

f2_ax4 = fig2.add_subplot(spec2[2, 2])
plt.title('[2, 2]') # plt.title()


f2_ax4 = fig2.add_subplot(spec2[3, 1])
plt.plot(a, a)
plt.plot(a, a**2)
plt.title('[3, 1]');
```

```python
# Indexes start from 0

fig3 = plt.figure(constrained_layout=True)

gs = fig3.add_gridspec(4, 4)

f3_ax1 = fig3.add_subplot(gs[0, :])
f3_ax1.set_title('[0, :]')
plt.plot(a, a)
plt.plot(a, a**2)


f3_ax2 = fig3.add_subplot(gs[1:, 0])
f3_ax2.set_title('[1:, 0]')

f3_ax3 = fig3.add_subplot(gs[1:3, 1:3])
plt.title('[1:3, 1:3]')
plt.plot(a, a)
plt.plot(a, a**2)

f3_ax4 = fig3.add_subplot(gs[1,3])
plt.title('[1, 3]')


f3_ax5 = fig3.add_subplot(gs[2, 3])
plt.title('[2, 3]')


f3_ax6 = fig3.add_subplot(gs[-1, 1:])
plt.title('[-1, 1:]');
```

### Different subplot

We can different subplot using gridspec or just plt.subplot(), below examples show tha same plots but differents way of codes, but using gridspec is more easier to shape different subplots

```python
fig4 = plt.figure(constrained_layout=True)
gs = fig4.add_gridspec(3, 3)

# ...
fig4.add_subplot(gs[0, :])
plt.title('[0, :]')
plt.plot(a, a)
plt.plot(a, a**2)


# ...
fig4.add_subplot(gs[1:, 0])
plt.title('[1:, 0]')
plt.plot(a, a)
plt.plot(a, a**2)


# ...
fig4.add_subplot(gs[1, 1:])
plt.title('[1, 1:]')
plt.plot(a, a)
plt.plot(a, a**2)


# ...
fig4.add_subplot(gs[2, 1])
plt.title('[2, 1]')
plt.plot(a, a)
plt.plot(a, a**2)


# ...
fig4.add_subplot(gs[2, 2])
plt.title('[2, 2]')
plt.plot(a, a)
plt.plot(a, a**2);
```

```python
plt.figure(constrained_layout=True)

# ...
plt.subplot(311)
plt.title('311')
plt.plot(a, a)
plt.plot(a, a**2)


# ...
plt.subplot(2,3,4)
plt.title('2,3,4')
plt.plot(a, a)
plt.plot(a, a**2)



# ...
plt.subplot(3,2,4)
plt.title('*')
plt.plot(a, a)
plt.plot(a, a**2)

# ...
plt.subplot(3,3,8)
plt.title('3,3,8')
plt.plot(a, a)
plt.plot(a, a**2)



# ...
plt.subplot(3,3,9)
plt.title('3,3,9')
plt.plot(a, a)
plt.plot(a, a**2)

```

### Fine Adjustments to a Gridspec Layout

```python
fig9 = plt.figure(figsize=(10,10), constrained_layout=False)

# ...
gs1 = fig9.add_gridspec(nrows=3, ncols=3, left=0.05, right=0.38,
                        wspace=0.5)
f9_ax1 = fig9.add_subplot(gs1[:-1, :])
f9_ax2 = fig9.add_subplot(gs1[-1, :-1])
f9_ax3 = fig9.add_subplot(gs1[-1, -1])

# ...
gs2 = fig9.add_gridspec(nrows=3, ncols=3, left=0.5, right=0.85,
                        hspace=0.5)
f9_ax4 = fig9.add_subplot(gs2[:-1, :])
f9_ax5 = fig9.add_subplot(gs2[-1, :-1])
f9_ax6 = fig9.add_subplot(gs2[-1, -1])
```
