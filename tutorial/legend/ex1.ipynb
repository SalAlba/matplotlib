*bbox_to_anchor & loc*

When `bbox_to_anchor` and loc are used together, the `loc` argument will inform matplotlib which part of the bounding box of the legend should be placed at the arguments of bbox_to_anchor. For example (I've simplified the command a bit), the three options below will produce different locations for your legend,

```python
plt.figure(figsize=(10,10), dpi=80)

# ...
plt.subplot(221)
plt.title('221')
plt.plot(a, a**2, 'r.', label='line 1')
plt.plot(a, a**3, 'b+', label='line 2')
plt.legend(bbox_to_anchor=[0.5, 0.5], loc='center');

# ...
plt.subplot(222)
plt.title('222')
plt.plot(a, a**2, 'r.', label='line 1')
plt.plot(a, a**3, 'b+', label='line 2')
plt.legend(bbox_to_anchor=[0.5, 0.5], loc='center left');

# ...
plt.subplot(223)
plt.title('223')
plt.plot(a, a**2, 'r.', label='line 1')
plt.plot(a, a**3, 'b+', label='line 2')
plt.legend(bbox_to_anchor=[0.5, 0.5], loc='center right');
```

```python
plt.figure(figsize=(10,10), dpi=80)

# ...
plt.subplot(221)
plt.title('221')
plt.plot(a, a**2, 'r.', label='line 1')
plt.plot(a, a**3, 'b+', label='line 2')
plt.legend(
    bbox_to_anchor=(0., 1.05, 1., .102),
    loc='center',
    ncol=2,
    mode="expand",
    borderaxespad=0.
);

# ...
plt.subplot(222)
plt.title('221')
plt.plot(a, a**2, 'r.', label='line 1')
plt.plot(a, a**3, 'b+', label='line 2')
plt.legend(
    bbox_to_anchor=(1.02, .02, 1., 1.1),
#     ncol=2,
    loc='center',
    mode="expand",
    borderaxespad=0.
);
```

```python
plt.figure(figsize=(10,10), dpi=80)

# ...
plt.subplot(211)
plt.title('221')
plt.plot(a, a**2, 'rx', label='line 1')
plt.plot(a, a**3, 'bo', label='line 2')
plt.legend();

# ...
plt.subplot(224)
plt.title('221')
plt.plot(a, a**2, 'r:', label='line 1')
plt.plot(a, a**3, 'b+', label='line 2')

plt.legend();
```

### Multiple legends on the same Axes

```python
# ...
line1, = plt.plot([1, 2, 3], label="Line 1", linestyle='--')
line2, = plt.plot([3, 2, 1], label="Line 2", linewidth=4)

# Create a legend for the first line.
first_legend = plt.legend(handles=[line1], loc='upper right')

# Create another legend for the second line.
second_legend = plt.legend(handles=[line2], loc='lower right')

# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)


plt.show()
```
